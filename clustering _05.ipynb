# Let's create sample dataset

```python
import numpy as np
from sklearn.datasets import make_classification
from matplotlib import pyplot
```

```python
X, y = make_classification(n_samples = 1201, n_features=2, n_informative = 2, n_redundant = 0, n_clusters_per_class = 1, random_state = 4)
```

```python
X
```

```python
y
```

```python
for class_value in range(2):
    row_ix = np.where(y == class_value)
    pyplot.scatter(X[row_ix,0], X[row_ix, 1])
pyplot.show()
```

# Let's apply different clustering algorithms

There are many clustering algorithms, lets explore few of them.

# 1) Affinity Propogation

Involves finding a set of exemplars that best summarize the data

```python
from sklearn.cluster import AffinityPropagation
model = AffinityPropagation(damping = 0.9)

# If you are trying in individual file, 
#   than you have to import more libraries
#   which are imported in above sections
```

Let's fit affinity propogation model into our dataset and predict

```python
model.fit(X)
yhat = model.predict(X)
```

```python
clusters = np.unique(yhat)
clusters
```

```python
for cluster in clusters:
    row_ix = np.where(yhat == cluster)
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
pyplot.show()
```

# 2) Agglomerative

involves merging examples until the desired number

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=5)
yhat = model.fit_predict(X)
```

```python
clusters = np.unique(yhat)
```

```python
for cluster in clusters:
    row_ix = np.where(yhat == cluster)
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
pyplot.show()
```

# 3) BIRCH

involves constructing a tree structure from which cluster centroids are extracted

```python
from sklearn.cluster import Birch
```

```python
model = Birch(threshold=0.01, n_clusters=3)
model.fit(X)
yhat = model.predict(X)
```

```python
clusters = np.unique(yhat)
clusters
```

```python
for cluster in clusters:
    a = np.where(yhat==cluster)
    pyplot.scatter(X[a,0], X[a,1])
pyplot.show()
```

# 4) DBSCAN

density-based notion of clusters which is designed to discover clusters of arbitrary shape

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps=0.30, min_samples=9)
yhat = model.fit_predict(X)
clusters = np.unique(yhat)
clusters
```

```python
for cluster in clusters:
    d = np.where(yhat == cluster)
    pyplot.scatter(X[d,0], X[d,1])
pyplot.show()
```

# 5) K-Means

N-dimensional population into k sets on the basis of a sample

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters= 5)
kmeans.fit(X)
yhat = kmeans.predict(X)
```

```python
kclusters = np.unique(yhat)
```

```python
for cluster in kclusters:
    kcl = np.where(cluster == yhat)
    pyplot.scatter(X[kcl,0], X[kcl,1])
```
